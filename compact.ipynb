```python
import pandas as pd
df2023 = pd.read_csv('survey_results_public.csv')
df2022 = pd.read_csv("2022.csv")
df2021 = pd.read_csv("2021.csv")
df2020 = pd.read_csv("2020.csv")
df2019 = pd.read_csv("2019.csv")

```

```python
df2023 = df2023[['LanguageHaveWorkedWith','LanguageWantToWorkWith']].dropna()
df2022 = df2022[['LanguageHaveWorkedWith','LanguageWantToWorkWith']].dropna()
df2021 = df2021[['LanguageHaveWorkedWith','LanguageWantToWorkWith']].dropna()
df2020 = df2020[['LanguageWorkedWith','LanguageDesireNextYear']].dropna()
df2019 = df2019[['LanguageWorkedWith','LanguageDesireNextYear']].dropna()

df2023.rename(columns={'LanguageHaveWorkedWith': 'LanguageHaveWorkedWith2023','LanguageWantToWorkWith': 'LanguageWantToWorkWith2023'},inplace=True)
df2022.rename(columns={'LanguageHaveWorkedWith': 'LanguageHaveWorkedWith2022','LanguageWantToWorkWith': 'LanguageWantToWorkWith2022'},inplace=True)
df2021.rename(columns={'LanguageHaveWorkedWith': 'LanguageHaveWorkedWith2021','LanguageWantToWorkWith': 'LanguageWantToWorkWith2021'},inplace=True)
df2020.rename(columns={'LanguageWorkedWith': 'LanguageHaveWorkedWith2020','LanguageDesireNextYear': 'LanguageWantToWorkWith2020'},inplace=True)
df2019.rename(columns={'LanguageWorkedWith': 'LanguageHaveWorkedWith2019','LanguageDesireNextYear': 'LanguageWantToWorkWith2019'},inplace=True)
```

```python
df = pd.concat([df2023,df2022,df2021,df2020,df2019],axis=1)
df
```

```python
df.to_csv('languages.csv',index=False)
```
